```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/hp/Downloads/customer_shopping_behavior.csv")
```

```python
# cleaning column names
import re
def clean_columns(col: str):
    col = col.strip().lower()
    col = col.replace(' ','_')
    col = re.sub(r'[^a-z0-9_]','',col)
    return col

df.columns = [clean_columns(c) for c in df.columns]
df.rename(columns = {'purchase_amount_usd': 'sales'},inplace=True)
```

```python
df.describe(include='all')
```

```python
#check null values
df.isnull().sum()
```

## Replacing null values by the median of the review ratings grouped by category

```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

## create a new column age_group

```python
labels = ['young adult','adult','middle age','senior']
age_bins = np.linspace(min(df['age']),max(df['age']),5)
df['age_group'] = pd.cut(df['age'],bins=age_bins,labels=labels,include_lowest=True)

#labels = ['young adult','adult','middle age','senior']
#df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
# create column purchase_frequency_days

frequency_mapping = {'Fortnightly':14, 
                     'Weekly':7,
                     'Annually':365,
                     'Quarterly':90,
                     'Bi-Weekly':14,
                     'Monthly':30, 
                     'Every 3 Months':90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
